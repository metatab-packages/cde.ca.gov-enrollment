```python
import seaborn as sns
import metapack as mp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display 

%matplotlib inline
sns.set_context('notebook')
mp.jupyter.init()
```

```python
#pkg = mp.jupyter.open_package()
pkg = mp.jupyter.open_source_package()
pkg
```

```python
def assign_year(n):
    _, y = n.split('_')
    return int(y)

e_frames = [r.dataframe().assign(year=assign_year(r.name)) for r in pkg.references() if r.name.startswith('enrollment') ]
```

```python
df = pd.concat(e_frames)

columns = [df.columns[-1]] + list(df.columns[:-1]) # Move 'year' to the beginning

df= df[columns]
```
