```python
import seaborn as sns
import metapack as mp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display 

%matplotlib inline
sns.set_context('notebook')
mp.jupyter.init()
```

```python
pkg = mp.jupyter.open_package() # Note! Working off built package!
#pkg = mp.jupyter.open_source_package()
pkg
```

```python
enrollment = pkg.resource('enrollment').dataframe()
```

```python
col_map = {
    0:'na',
    1:'aian',
    2:'asian',
    3:'pi',
    4:'filipino',
    5:'hisp',
    6:'black',
    7:'nh_white',
    8:'unk', # Not in data dictinoary
    9:'two_more'
}

t = enrollment[['cdscode', 'ethnic','year','enr_total']]
raceeth = t.groupby(['cdscode', 'year','ethnic']).sum().unstack()

raceeth.columns = [col_map.get(e,e) for e in raceeth.columns.levels[1]]
raceeth = raceeth.rename({'cdscode':'cdscode'})
raceeth['total'] = raceeth.sum(axis=1)


raceeth = raceeth.fillna(0).astype(int)
raceeth.head(20)
```
